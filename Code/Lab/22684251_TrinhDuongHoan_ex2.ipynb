## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

```python
import numpy as np
arr = np.random.randint(0, 1000, 100)
```

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
print(arr[:10])
```

```python
print(arr[-5:])
```

```python
for val in arr :
    if val % 5 == 0 and val > 100 :
        print(val, end = ' ')
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
less_than_100 = [int(val) for val in arr if val < 100]
print(len(less_than_100))
print(less_than_100)
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
# Tổng mảng
print(np.sum(arr))
```

```python
# Giá trị trung bình
print(np.mean(arr))
```

```python
# Giá trị trung vị
print(np.median(arr))
```

```python
# Q1, Q2, Q3
print(np.percentile(arr, 25))
print(np.percentile(arr, 50))
print(np.percentile(arr, 75))
```

```python
# Phân vị 20 
print(np.percentile(arr, 20))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
for i in range (10) :
    select_10 = np.random.choice(arr, 10)
    print(f"Lần chọn thứ {i + 1} : {select_10}")
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
from collections import Counter
count = Counter(arr)
print("Phần tử xuất hiện nhiều nhất : ", count.most_common(1)[0][0])
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
# y = sin(x)
pi = np.pi
x = np.linspace(0, 2 * pi, 100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'sin(x)', color = 'blue')
plt.plot(x, y2, linestyle = '--' , label = 'cos(x)', color = 'red')
plt.grid()
plt.legend()
plt.show()
```

### Câu 5: Vẽ các đồ thị con

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)
plt.title("Đồ thị hàm sin(x) và cos(x)")
plt.plot(x, y1, label = 'sin(x)', color = 'blue')
plt.grid()
plt.legend()
plt.subplot(2, 1, 2)
plt.plot(x, y2, label = 'cos(x)', color = 'red')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

```python
# This cell contains output. Don't remove!
```

```python
ar = np.random.normal(size=1000)
plt.subplot(2, 1, 1)
plt.hist(ar, bins = 100, color='#0066FF', edgecolor='white')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.subplot(2, 1, 2)
plt.boxplot(ar, vert = False)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Boxplot')
plt.tight_layout()
plt.show()
```
